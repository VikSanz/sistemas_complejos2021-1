```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 6]
from re import sub
from collections import defaultdict, Counter
```

```python
import elotl.corpus
## El método corpus.load() nos regresa una lista multidimensional. 
## Cada elemento de esta lista corresponde a una oración paralela del corpus más información extra del par de oraciones
corpus1=elotl.corpus.load('axolotl')  #español-náhuatl 
corpus2=elotl.corpus.load('tsunkua')  #español-otomi 
```

```python
### El corpus está formado por oraciones paralelas entre ambas lenguas. i.e. Nahuatl-español y Otomí-español.
### Para acceder a ellos toma en cuenta que: [oracion_en_el_corpus1][0=español,1=nahuatl]
corpus1[0][1]
```

```python
corpus1[1][1]
```

```python
### El corpus está formado por oraciones paralelas entre ambas lenguas. i.e. Nahuatl-español y Otomí-español.
### [oracion_en_el_corpus1][0=español,1=nahuatl]
corpus1[0][0]
```

```python
corpus1[1][0]
```

```python
### El corpus está formado por oraciones paralelas entre ambas lenguas. i.e. Nahuatl-español y Otomí-español.
### [oracion_en_el_corpus2][0=español,1=otomí]
corpus2[1][1]
```

```python
### Vamos a extraer la lista de palabras para cada lengua en cada corpus: 
corpus1_es=[]
corpus1_na=[] 
palabras=[] ### Lista temporal

### Iteramos sobre cada línea de los corpus y vamos guardando las palabras que corresponden a cada lengua:
###Corpus español-náhuatl
for row in corpus1:   
  ### row[0]: corresponde con las oraciones en español
  row[0]=sub(r'[^\w\s]',' ',row[0])     #Sustituimos signos de puntuación por un espacio
  palabras=row[0].lower().split()       #Dividimos a la oración en palabras (el delimitador es un espacio). Luego, guardamos cada una de estas palabras en la lista palabras
  for p in palabras:                    #Iteramos sobre cada palabra de la oración (lista palabras)
       corpus1_es.append(p)             #vamos agregando cada palabra a una lista (corpus1_es)
  ### row[1]: corresponde con las oraciones en nahuatl   
  row[1]=sub(r'[^\w\s\']',' ',row[1])  #Sustituimo signos de puntuación por un espacio. Notemos que aparece este signo "'", también lo quitamos.
  palabras=row[1].lower().split()      #Dividimos a la oración en palabras (el delimitador es un espacio). Luego, guardamos cada una de estas palabras en la lista palabras
  for p in palabras:                      #Iteramos sobre cada palabra de la oración (lista palabras)
       corpus1_na.append(p)            #vamos agregando cada palabra a una lista (corpus1_na)
```

```python
len(corpus1_na)
```

```python
### Notemos que hay más palabras en español que en nahuatl: esto es porque el nahuatl es una lengua aglutinante.
len(corpus1_es)
```

```python
### Vamos a extraer la lista de palabras para cada lengua en cada corpus: 
corpus2_es=[]
corpus2_ot=[] 
palabras=[] ### Lista temporal

### Iteramos sobre cada línea de los corpus y vamos guardando las palabras que corresponden a cada lengua:
###Corpus español-otomí
for row in corpus2:   
  ### row[0]: corresponde con las oraciones en español
  row[0]=sub(r'[^\w\s]',' ',row[0])     #Sustituimos signos de puntuación por un espacio
  palabras=row[0].lower().split()       #Dividimos a la oración en palabras (el delimitador es un espacio). Luego, guardamos cada una de estas palabras en la lista palabras
  for p in palabras:                       #Iteramos sobre cada palabra de la oración (lista palabras)
       corpus2_es.append(p)             #vamos agregando cada palabra a una lista (corpus2_es)
  ### row[1]: corresponde con las oraciones en otomí   
  row[1]=sub(r'[^\w\s\']',' ',row[1])  #Sustituimo signos de puntuación por un espacio. Notemos que aparece este signo "'", también lo quitamos.
  palabras=row[1].lower().split()      #Dividimos a la oración en palabras (el delimitador es un espacio). Luego, guardamos cada una de estas palabras en la lista palabras
  for p in palabras:                      #Iteramos sobre cada palabra de la oración (lista palabras)
       corpus2_ot.append(p)            #vamos agregando cada palabra a una lista (corpus2_ot)
```

```python
corpus1_na
```

```python
### Hay palabras aún más curiosas y con signos de puntuación curiosos: dínkhák'ŭ, ngŭ, k'ehña
corpus2_ot
```

```python
### La función Counter() nos regresará un tipo  diccionario donde cada elemento es un par: (palabra, frecuencia)
corpus1_na_voc=Counter(corpus1_na)
corpus1_es_voc=Counter(corpus1_es)
corpus2_ot_voc=Counter(corpus2_ot)
corpus2_es_voc=Counter(corpus2_es)
```

```python
### Palabras más comunes del Nahuatl
corpus1_na_voc.most_common(10)
```
