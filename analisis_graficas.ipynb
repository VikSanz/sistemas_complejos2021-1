```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import os
import graficas_ayuda as gr
```

```python
t = "metro-metrobus"
G = nx.read_gexf(os.path.join("transporte","{0}.gexf".format(t)))
```

```python
degs = ["degree","in_degree","out_degree"]
for deg in degs:
    func = getattr(G,deg)
    degreeSequence = sorted([d for n,d in func()])
    m1 = degreeSequence[0]
    m2 = degreeSequence[-1]
    if m1==0:
        m1=1
    plt.figure()
    plt.hist(degreeSequence,alpha=0.7,ec="k")
    plt.xlabel(deg)
    plt.ylabel("Frequency")
    plt.title(t)
    plt.grid()
    plt.show()
```

```python
centrals = [nx.centrality.degree_centrality,
            nx.centrality.betweenness_centrality,
            nx.centrality.closeness_centrality,
           nx.pagerank]
for func in centrals:
    plt.figure()
    xs = func(G).values()
    plt.hist(xs,bins=20,alpha=0.6,ec="k")
    plt.title(t)
    xlab = func.__name__
    plt.xlabel(xlab)
    plt.ylabel("Frequency")
    plt.grid()
    plt.show()
```

```python
centrals = [nx.centrality.degree_centrality,
            nx.centrality.betweenness_centrality,
            nx.centrality.closeness_centrality,
           nx.pagerank]
for func in centrals:
    xs = func(G)
    print(func.__name__)
    orden =[(k,v) for k,v in sorted(xs.items(),key=lambda item:(1-item[1]))]
    for i in range(10):
        print(orden[i])
    print()
```
